```python
import pandas as pd
df=pd.read_csv("weather_data.csv",parse_dates=["day"]) #changes the date format to valid format yyyy-mm-dd
df.set_index('day',inplace=True)
df
```

```python
df.isnull().sum()
```

```python
new_df=df.fillna(0)
new_df
```

```python
#if wanted to fill the null value with different value for each column

new_df=df.fillna({
    'temoerature':0,
    'windspeed':1,
    'event':'snow'
})
new_df
#if there is a null value in bmi it will be set as '0', if in hypertension then set it as '1',ifelse in ever_marrried then set as "yes"
```

```python
#to set the null value with previous days datas or previous persons data rather than setting all null values as 0,1 etc...
new_df=df.fillna(method='ffill')
new_df
```

```python
#to set the null value with next days datas or next persons data rather than setting all null values as 0,1 etc...
new_df=df.fillna(method='bfill')
new_df
```

```python
new_df=df.fillna(method='bfill',axis='columns')
new_df
#this replace the null value horizontally
```

```python
new_df=df.fillna(method='ffill',limit=1)
new_df
```

```python
new_df=df.fillna(method='ffill',limit=2)
new_df
```

```python
new_df=df.interpolate()
new_df
#it replaces the null value with average of above and below data
```

```python
new_df=df.interpolate(method="time")
new_df
#this replace the null values with avg of nearer date
```

```python
df
```

```python
new_df=df.dropna()
new_df
```

```python
new_df=df.dropna(how="all")
new_df
#this removes the row completely only if all the elements in the rows are null
```

```python
new_df=df.dropna(thresh=2)
new_df
#this will NOT remove the row if there are atleast 2 valid values in the row
```

```python
#inserting missing dates
dt=pd.date_range("01-01-2017","01-11-2017")
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```
